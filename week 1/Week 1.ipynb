```python
import gradio as gr
```

```python
def chatbot(message, history):

    message = message.lower().strip()

    if message in ["hi", "hello", "hey"]:
        reply = "👋 Hello! Welcome."

    elif message in ["how are you", "how are you?"]:
        reply = "😊 I'm doing great."

    elif message == "what is ai":
        reply = "Artificial Intelligence enables machines to think and solve problems."

    elif message == "python":
        reply = "Python is a popular programming language."

    elif message == "machine learning":
        reply = "Machine Learning allows computers to learn from data."

    elif message == "deep learning":
        reply = "Deep Learning is a subset of Machine Learning."

    elif message == "who made you":
        reply = "I was created using Python."

    elif message == "what is your name":
        reply = "My name is RuleBot."

    elif message in ["bye", "exit", "quit"]:
        reply = "👋 Goodbye!"

    else:
        reply = "Sorry, I don't understand."

    history.append({"role":"user","content":message})
    history.append({"role":"assistant","content":reply})

    return "", history
```

```python
with gr.Blocks() as demo:

    gr.Markdown("# 🤖 Rule-Based AI Chatbot")

    chatbot_ui = gr.Chatbot(height=450)

    msg = gr.Textbox(
        label="Your Message",
        placeholder="Type here..."
    )

    send = gr.Button("🚀 Send")
    clear = gr.Button("🗑️ Clear")

    send.click(
        chatbot,
        inputs=[msg, chatbot_ui],
        outputs=[msg, chatbot_ui]
    )

    msg.submit(
        chatbot,
        inputs=[msg, chatbot_ui],
        outputs=[msg, chatbot_ui]
    )

    clear.click(lambda: [], outputs=chatbot_ui)

demo.launch(theme=gr.themes.Soft())
```
